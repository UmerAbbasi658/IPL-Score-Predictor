```python
import pandas as pd
import numpy as np
import pickle
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv('/kaggle/input/ipl-data/ipl.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.drop(['mid','batsman','bowler','striker','non-striker','venue'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df.bat_team.unique()
```

```python
cons_teams=['Kolkata Knight Riders','Chennai Super Kings','Rajasthan Royals','Mumbai Indians','Kings XI Punjab','Royal Challengers Bangalore','Delhi Daredevils','Sunrisers Hyderabad','Rising Pune Supergiants']
```

```python
df=df[(df['bat_team'].isin(cons_teams))&(df['bowl_team'].isin(cons_teams))]
```

```python
df.head()
```

```python
df = df[df['overs'] >= 5]
```

```python
from datetime import datetime
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.dtypes
```

```python
encoded_df = pd.get_dummies(data=df, columns=['bat_team', 'bowl_team'], dtype=int)
```

```python
encoded_df.head()
```

```python
x_train=encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year<=2016]
x_test=encoded_df.drop(labels='total',axis=1)[encoded_df['date'].dt.year>=2017]
```

```python
y_train = encoded_df[encoded_df['date'].dt.year <= 2016]['total'].values
y_test = encoded_df[encoded_df['date'].dt.year >= 2017]['total'].values
```

```python
x_train.drop(labels='date',axis=True,inplace=True)
x_test.drop(labels='date',axis=1,inplace=True)
```

```python
x_train
```

```python
from sklearn.linear_model import LinearRegression
model=LinearRegression()
model.fit(x_train,y_train)
```

```python
from sklearn.metrics import r2_score
y_pred=model.predict(x_test)
```

```python
r2_score=r2_score(y_test,y_pred)
```

```python
r2_score
```

```python
pickle.dump(model,open('lr_model.pk1','wb'))
```
